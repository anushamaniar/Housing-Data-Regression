## Scaling and Train Test Split

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

**Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape #19 features so use 19 neurons
```

```python
X_test.shape
```

# Creating a model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse') #as regression use loss as mse
```

```python
model.fit(x=X_train,y=y_train.values,validation_data=(X_test,y_test.values),batch_size=128,epochs=400)
#use .values for creating numpy array instead of pd series/df as tensorflow doesnt work w pd that well
#validation_data used to check loss on test data, this test data wont affect weights/biases of our network
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#if orange line spikes up then that means overfitting
#but as it is not going up, we are good
```

# Evaluation on test data

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions) #mae
```

```python
np.sqrt(mean_squared_error(y_test,predictions)) #rmse
```

```python
explained_variance_score(y_test,predictions) #best - 1
```
